```python
import os
import cv2
import numpy as np

# Paths to the folders
train_masks_path = r'C:\Users\91741\Downloads\Implement\train_masks\train_masks'

# Limit to the first 100 images
limit = 100

# List all mask files and sort them to match the order used in training
mask_files = sorted(os.listdir(train_masks_path))[:limit]

# Counters for tumorous and non-tumorous masks
tumorous_count = 0
non_tumorous_count = 0

# List to keep track of which images are tumorous or non-tumorous
tumor_status = []

# Iterate through the 100 mask files
for mask_file in mask_files:
    mask_path = os.path.join(train_masks_path, mask_file)
    
    # Read the mask image in grayscale mode
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    
    # Check if there are any non-zero pixels in the mask
    if np.any(mask > 0):
        tumor_status.append((mask_file, 'Tumorous'))
        tumorous_count += 1
    else:
        tumor_status.append((mask_file, 'Non-Tumorous'))
        non_tumorous_count += 1

# Print the results
print(f"Tumorous masks: {tumorous_count}")
print(f"Non-tumorous masks: {non_tumorous_count}")

# Display the list of the first 100 images with their tumor status
for mask_file, status in tumor_status:
    print(f"{mask_file}: {status}")
```
