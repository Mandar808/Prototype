```python
import os
import cv2
import numpy as np

# Paths to the folders
train_images_path = r'C:\Users\91741\Downloads\Implement\train_images\train_images'
train_masks_path = r'C:\Users\91741\Downloads\Implement\train_masks\train_masks'

# Lists to store the verified images as tumorous or non-tumorous
tumorous_images = []
non_tumorous_images = []

# Counter to track how many images have been verified
verified_count = 0
target_count = 1000  # Limit to 1000 images (500 tumor and 500 non-tumor)

# Iterate over mask files and check each image
for mask_file in sorted(os.listdir(train_masks_path)):
    mask_path = os.path.join(train_masks_path, mask_file)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    
    # Check if mask has non-zero pixels (indicating a tumor)
    if np.any(mask > 0):
        if len(tumorous_images) < 500:
            tumorous_images.append(mask_file)
    else:
        if len(non_tumorous_images) < 500:
            non_tumorous_images.append(mask_file)
    
    # Stop if we've gathered 500 tumorous and 500 non-tumorous images
    verified_count = len(tumorous_images) + len(non_tumorous_images)
    if verified_count >= target_count:
        break

# Print the results
print("Verified Tumorous Images:")
print(tumorous_images)
print("\nVerified Non-Tumorous Images:")
print(non_tumorous_images)

print(f"\nTotal Tumorous Images Verified: {len(tumorous_images)}")
print(f"Total Non-Tumorous Images Verified: {len(non_tumorous_images)}")
```
